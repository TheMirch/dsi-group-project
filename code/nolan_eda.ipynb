```python
import pandas as pd
import matplotlib.pyplot as plt

# pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# pd.reset_option('display.max_rows')
# pd.reset_option('display.max_columns')
```

```python
ak = pd.read_csv('../datasets/alaska_single_engine_clean.csv', low_memory=False)
```

```python
ak.Model.value_counts()
```

```python
model_counts = ak.Model.value_counts().head(30)
```

```python
plt.figure(figsize=(12, 6))
model_counts.plot(kind='bar')
plt.title('Top 30 Aircraft Models vs. Number of Incidents')
plt.xlabel('Aircraft Model')
plt.ylabel('Incidents')
plt.show()
```

```python
ak.HighestInjuryLevel.value_counts()
```

```python
injury_level_counts = ak.HighestInjuryLevel.value_counts()
```

```python
plt.figure(figsize=(12, 6))
injury_level_counts.plot(kind='bar')
plt.title('Injury Level vs. Number of Incidents')
plt.xlabel('Injury Level')
plt.ylabel('Incidents')
plt.savefig('../images/incidents_by_injury_level.png')
plt.show()
```

```python
ak.AirCraftDamage.value_counts()
```

```python
aircraft_damage_counts = ak.AirCraftDamage.value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
aircraft_damage_counts.plot(kind='bar')
plt.title('Top 10 Aircraft Damage Types vs. Number of Incidents')
plt.xlabel('Aircraft Damage')
plt.ylabel('Incidents')
plt.show()
```

```python
ak.EventSeason.value_counts()
```

```python
season_counts = ak.EventSeason.value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
season_counts.plot(kind='bar')
plt.title('Incidents by Season')
plt.xlabel('Season')
plt.ylabel('Incidents')
plt.savefig('../images/incidents_by_season.png')
plt.show()
```
