```python
ak = pd.read_csv('../datasets/alaska_single_engine_clean_camel.csv', low_memory=False)
```

```python
ak.Model.value_counts()
```

```python
model_counts = ak.Model.value_counts().head(30)
```

```python
ak_num = ak.drop(['NtsbNo', 'Mkey', 'City', 'N', 'ProbableCause', 'Make', 'Model', 'AirportID', 'AirportName', 'FAR', 'Operator', 'EventTime',], axis=1)
```

```python
ak_num.head()
```

```python
categorical_cols = ak_num.select_dtypes(exclude=['number']).columns.tolist()
```

```python
categorical_cols
```

```python
ak_num = pd.get_dummies(ak_num, columns = categorical_cols, drop_first=False)
```

```python
ak_num.head()
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(ak_num.corr(), cmap='coolwarm', fmt=".2f", vmin= -1, annot=False);
```

```python
plt.figure(figsize = (5, 20))
sns.heatmap(ak_num.corr()[['HighestInjuryLevel_Fatal', 'HighestInjuryLevel_Minor', 'HighestInjuryLevel_None Reported', 'HighestInjuryLevel_Serious']], cmap='coolwarm', fmt=".2f", vmin= -1, annot=True);
```
